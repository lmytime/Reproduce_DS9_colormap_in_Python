```python
from astropy.io import fits
from astromy_ds9 import ds9_norm

# Read the image
with fits.open('data/subaru-r.fits') as hdul:
    image = hdul[1].data
```


Image in DS9 looks like:

<img src="./figs/example1_ds9.jpg" alt="example1" width=500/>

```python
import matplotlib.pyplot as plt
import numpy as np

vmin, vmax = -0.0561133, 0.126864
bias, contrast = 0.66, 2.28

fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)

norm = ds9_norm(vmin=vmin, vmax=vmax, bias=bias, contrast=contrast, stretch='linear')
im = ax.imshow(image.data, cmap='Greys', origin='lower', norm=norm, interpolation='none')
plt.colorbar(im, ax=ax, boundaries=np.linspace(vmin, vmax, 1000))
plt.title('ds9-like image inspection with Matplotlib, linear stretch')
plt.savefig('figs/example1_py.png', dpi=200, bbox_inches='tight')
plt.show()
```
